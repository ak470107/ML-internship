## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*
## Research Question

### Search Question
Which content pages should be prioritized for review based on historical search and engagement data?

### Decision
Help content managers decide which pages should be reviewed first.

### Action
A content manager can refresh, expand, rewrite, merge, or monitor the pages that receive the highest priority scores.

### Unit of Analysis
One content page.

### Output
A ranked list of pages with a priority score and reason codes explaining why each page was recommended.

### Cost of a Wrong Recommendation
If the recommendations are incorrect, time may be spent reviewing pages that do not need attention while important pages are missed. This reduces the efficiency of the content review process.

### Why Machine Learning?
Historical search and engagement metrics contain patterns that are difficult to identify manually across many pages. Machine learning can help prioritize pages that appear most likely to require review, while the final decision remains with a human reviewer.
